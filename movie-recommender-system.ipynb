```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#genres
#id (to fetch posters of movies)
#keywords
#title
#overview
#cast
#crew

# extract important data from data set
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
# check missing entries
movies.isnull().sum()
```

```python
# replace/remove missing entries
movies.dropna(inplace=True)
```

```python
# check duplicate entries
movies.duplicated().sum()
```

```python
import ast
```

```python
#pre processing function
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): 
        #ast.literal_eval() converts string to python dictionary object/list
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#pre processing function + to get top data from first 3 dictionaries only in string
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj): 
        #ast.literal_eval() converts string to python dictionary object/list
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
#pre processing function to get director from crew coloumn as only director is important while recommeding movies
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj): 
        #ast.literal_eval() converts string to python dictionary object/list
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] =movies['crew'].apply(fetch_director)
```

```python
#convert overview to list from string
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#to remove space and make one tag of multiple words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#form one single coloumn named tags which is concatenation of overview, genres, keywords, cast, crew
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
# create new data frame that only includes title, id and tags
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
# convert list to string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#convert tags to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Stemming is a technique used to reduce an inflected word down to its word stem.
# It helps removing morphological variation of a word's original root form and treat as one
```

```python
# check for inflected words iterating through every word in text
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
#     after stemming we convert to string and return
```

```python
new_df['tags'].apply(stem)
```

```python
# count vectorizer will select most used words in all tags and convert them into vector form
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
# max_features=5000, signify 5000 most used words will be considered, and stop_words='english' signify 
# that stop words of english language are to be ignored (stop words = a,an,the,are...etc.)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# at higher dimensions eucledian distance isnt a good indicator of similarity so we consider cosine distance.
# cosine similarity is basically inverse of cosine distance
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
#     enumerate is used to hold distances and index number together while sorting to prevent mixing

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
